Make request using 'search/tweets' api, parameters are attached in a dictionary.

```python
#function to clean the tweets
def cleaning_tweet(text):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|RT", " ", text).split())

#function to get the sentiment of a tweet
def tweet_sentiment(text):         
        analysis = TextBlob(text)         
        if analysis.sentiment.polarity > 0: 
            return 'positive'
        elif analysis.sentiment.polarity == 0: 
            return 'neutral'
        else: 
            return 'negative'
        
        
#function to fetch tweets for 'UberEats'
def get_tweets_uber(r):
# open an empty file for output
    with open ('UberEats.txt','a') as file1:
        c=0      
        for item in r:
            try:
              # index the output
                c += 1
                print(c)              
              # parse data to fetch various details   

                if 'screen_name' in item['user']:
                    if len(item['user']['screen_name'])>0:
                        screen_name=item['user']['screen_name']  
                    else:
                        screen_name='MISSING'
                if 'text' in item:
                    if len(item['text'])>0:
                        text= item['text']
                        #text=text.replace('\n',' ') 
                        text=cleaning_tweet(text)
                        sent=tweet_sentiment(text)
                    else:
                        text='MISSING'
                if 'id' in item:     
                    id1 = item['id']
                    if len(str(id1))!=0:
                        ids_uber.append(id1)
                    else:
                        id1='null'                                  
                if 'location' in item['user']:
                    if len(item['user']['location'])>0:
                        location=item['user']['location']  
                    else:
                        location='MISSING'
                if 'followers_count' in item['user']:
                    followers_count=item['user']['followers_count']
                if 'favourites_count' in item['user']:
                    favourites_count=item['user']['favourites_count']
                if 'retweet_count' in item:
                    retweet_count=item['retweet_count']
                if 'verified' in item['user']:
                    if item['user']['verified']=='True':
                        verified='Yes'
                    else:
                        verified='No'
                
                    
                print(screen_name,':',text,':',id1,':',location,':',followers_count,':',favourites_count,':',retweet_count,':',verified,':',sent)
              # write parsed data into a .txt file names Twitter_data.txt     
                file1.write(str(screen_name)+'\t'+str(text)+'\t'+str(id1)+'\t'+str(location)+'\t'+str(followers_count)+'\t'+str(favourites_count)+'\t'+str(retweet_count)+'\t'+str(verified)+'\t'+str(sent)+'\n')
            except UnicodeEncodeError:
                                pass

#function to fetch tweets for 'GrubHub'
def get_tweets_grubhub(r1):
    with open ('Grubhub.txt','a') as file2:
        c1=0
        # iterate through response r
        
        for item in r1:
            try:
              # index the output
                c1 += 1
                print(c1)                
              # parse data to fetch various details
                if 'screen_name' in item['user']:
                    if len(item['user']['screen_name'])>0:
                        screen_name=item['user']['screen_name']  
                    else:
                        screen_name='MISSING'
                if 'text' in item:
                    if len(item['text'])>0:
                        text= item['text'] 
                        #text=text.replace('\n',' ') 
                        text=cleaning_tweet(text)     
                        sent=tweet_sentiment(text)
                    else:
                        text='MISSING'
                if 'id' in item:  
                    id1 = item['id']
                    if len(str(id1))!=0:
                        ids_grubhub.append(id1)
                    else:
                        id1='null'                                  
                if 'location' in item['user']:
                    if len(item['user']['location']  )>0:
                        location=item['user']['location']  
                    else:
                        location='MISSING'
                if 'followers_count' in item['user']:
                    followers_count=item['user']['followers_count']
                if 'favourites_count' in item['user']:
                    favourites_count=item['user']['favourites_count']
                if 'retweet_count' in item:
                    retweet_count=item['retweet_count']
                if 'verified' in item['user']:
                    if item['user']['verified']=='True':
                        verified='Yes'
                    else:
                        verified='No'
                print(screen_name,':',text,':',id1,':',location,':',followers_count,':',favourites_count,':',retweet_count,':',verified,':',sent)                 
              # write parsed data into a .txt file names Twitter_data.txt     
                file2.write(str(screen_name)+'\t'+str(text)+'\t'+str(id1)+'\t'+str(location)+'\t'+str(followers_count)+'\t'+str(favourites_count)+'\t'+str(retweet_count)+'\t'+str(verified)+'\t'+str(sent)+'\n')
            except UnicodeEncodeError:
                                pass

#first run 
r = api.request('search/tweets', {'q': SEARCH_TERM, 'count': COUNT,'lang':LANG,'geocode':GEOCODE})
get_tweets_uber(r)
min1=min(ids_uber) 
#print(min1)
max1=int(min1)-1
max1=str(max1)
#loop to get next 100 tweets with max id being the min from last run
for i in range(1,30):
    r = api.request('search/tweets', {'q': SEARCH_TERM,'count':COUNT,'max_id':max1,'lang':LANG,'geocode':GEOCODE})
    get_tweets_uber(r)
    min1=min(ids_uber)
    max1=int(min1)-1
    max1=str(max1)
    
r1=api.request('search/tweets', {'q': SEARCH_TERM2, 'count': COUNT,'lang':LANG,'geocode':GEOCODE})
get_tweets_grubhub(r1)
print(ids_grubhub)
min2=min(ids_grubhub) 
#print(min1)
max2=int(min2)-1
max2=str(max2)
for i in range(1,30):
    r1 = api.request('search/tweets', {'q': SEARCH_TERM2,'count':COUNT,'max_id':max2,'lang':LANG,'geocode':GEOCODE})
    get_tweets_grubhub(r1)
    min2=min(ids_grubhub)
    max2=int(min2)-1
    max2=str(max2)
    
```
